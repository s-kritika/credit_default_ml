```python
import joblib
from joblib import dump, load
import numpy as np
import pandas as pd

# Load the models back from the file
logi = joblib.load('cred_model_l.csv')
rfc = joblib.load('cred_model_r.csv')
xgb = joblib.load('cred_model_x.csv')
```

```python
import numpy as np
from scipy.stats import mode
val_data = pd.read_csv("val_data.csv")
val_data=val_data.drop('account_number', axis=1)
```

```python
missing_data = val_data.isna().sum()
print(missing_data[missing_data == len(val_data)])  # Features with all missing values
val_data = val_data.dropna(axis=1, how='all')  
missing_data = val_data.isna().sum()
print(missing_data[missing_data == len(val_data)])  # Features with all missing values
print(val_data.shape)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
my_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('scalar', StandardScaler())
])
print(val_data.shape) 
df_imputed = my_pipe.fit_transform(val_data)
print(type(df_imputed))  # Should be <class 'numpy.ndarray'>
print(df_imputed.shape) 
df_imputed = pd.DataFrame(df_imputed, columns=val_data.columns)
#print(type(df_imputed)) 
print(df_imputed.columns)
```

```python
import pandas as pd
from scipy.stats import mode
import numpy as np

# Assuming logi, rfc, xgb are trained models, and df_imputed is your input DataFrame

# Get predicted probabilities from each model (for class 1)
prob1 = logi.predict_proba(df_imputed)[:, 1]  # Probabilities for class 1 (Logistic Regression)
prob2 = rfc.predict_proba(df_imputed)[:, 1]   # Probabilities for class 1 (Random Forest)
prob3 = xgb.predict_proba(df_imputed)[:, 1]   # Probabilities for class 1 (XGBoost)

# Create a list to store the mode probabilities
mode_probabilities = []

# Loop through each sample (row) and calculate the mode of the probabilities
for i in range(len(df_imputed)):  # Iterate through each sample
    # Get the predicted probabilities for the current sample from all three models
    preds = [prob1[i], prob2[i], prob3[i]]
    
    # Using scipy.stats.mode to find the most frequent probability
    mode_prob, _ = mode(preds)  # Get the mode of the probabilities
    
    # Append the mode probability (mode_prob is a scalar value)
    if isinstance(mode_prob, np.ndarray):
        mode_probabilities.append(round(mode_prob[0],2))  # If mode_prob is an array, take the first element
    else:
        mode_probabilities.append(round(mode_prob,2))
# Convert mode_probabilities to a DataFrame
df_mode_probabilities = pd.DataFrame({
    "Mode_Probabilities": mode_probabilities
})

# Save the mode probabilities to a CSV file
df_mode_probabilities.to_csv("final_probabilities.csv", index=False)

print("Mode probabilities have been saved to 'final_probabilities.csv'.")
```
