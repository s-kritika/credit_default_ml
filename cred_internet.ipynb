```python
dev_data=dev_data.drop('account_number', axis=1)
print(dev_data.shape)
```

```python
missing_data = dev_data.isna().sum()
print(missing_data[missing_data == len(dev_data)])  # Features with all missing values
dev_data = dev_data.dropna(axis=1, how='all')  
missing_data = dev_data.isna().sum()
print(missing_data[missing_data == len(dev_data)])  # Features with all missing values
print(dev_data.shape)
```

```python
print(dev_data.columns)  # To check if 'bad_flag' is actually present
```

```python
X = dev_data.drop('bad_flag' ,axis = 1)
y = dev_data['bad_flag'].copy()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
my_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('scalar', StandardScaler())
])
print(X.shape) 
df_imputed = my_pipe.fit_transform(X)
print(type(df_imputed))  # Should be <class 'numpy.ndarray'>
print(df_imputed.shape) 
df_imputed = pd.DataFrame(df_imputed, columns=X.columns)
#print(type(df_imputed)) 
print(df_imputed.columns)
```

```python
print(y.value_counts())
print(y.dtype)
```

```python
df_imputed.head()
```

```python
#importing SMOTE to handle class imbalance
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(df_imputed[(i for i in list(df_imputed.describe(include='all').columns) if i != 'bad_flag')], y)

print('Original unbalanced dataset shape', len(df_imputed))
print('Resampled balanced dataset shape', len(y_smote))
```

```python
print(y_smote.value_counts())
```

```python
print(type(x_smote))
x_smote=x_smote.copy()
```

```python
#creating new dataframe from balanced dataset after SMOTE
balanced_df = pd.DataFrame(x_smote, columns=list(i for i in list(df_imputed.describe(include='all').columns) if i != 'bad_flag'))
#adding target variable to new created dataframe
balanced_df['bad_flag'] = y_smote
#checking for class imbalance
plt.figure(figsize=(5,5))
sns.countplot(x ='bad_flag', data = balanced_df, palette = "Set1" )
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(balanced_df, test_size=0.3, random_state=42)
print(f" Rows in train set: {len(train_set)}\n Rows in test set: {len(test_set)}\n")
dev_data = train_set.copy()
```

```python
X_bal = train_set.drop('bad_flag',axis = 1)
y_bal = dev_data['bad_flag']
X_test = test_set.drop('bad_flag',axis = 1)
y_test = test_set['bad_flag']
```

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, log_loss
def metric_score(ac_value, prob_value):
    #evaluating training model
    
    pred_labels = (prob_value > 0.5).astype(int)

    model_train_accuracy = accuracy_score(ac_value, pred_labels)
    model_train_f1 = f1_score(ac_value, pred_labels, average='weighted')
    model_train_precision = precision_score(ac_value, pred_labels, zero_division=0)
    model_train_recall = recall_score(ac_value, pred_labels)
    model_train_rocauc_score = roc_auc_score(ac_value, pred_labels)
    log_loss_score = log_loss(ac_value, pred_labels, labels=[0, 1])


    print("accuracy: {:.4f}".format(model_train_accuracy))
    print("f1 score: {:.4f}".format(model_train_f1))
    print("precision: {:.4f}".format(model_train_precision))
    print("recall: {:.4f}".format(model_train_recall))
    print("roc_auc_score: {:.4f}".format(model_train_rocauc_score))
    print("log_loss: {:.4f}\n".format(log_loss_score))
    print(f"Classification Report:\n{classification_report(ac_value, pred_labels)}")
    print(f"Confusion Matrix:\n{confusion_matrix(ac_value, pred_labels)}")
    print("-" * 50)
```

```python
from sklearn.linear_model import LogisticRegression
logi = LogisticRegression(fit_intercept=True, max_iter=10000)
logi.fit(X_bal, y_bal)
y_pred_logi = logi.predict(X_bal)
print("model performance for training set: \n")
metric_score(y_bal, y_pred_logi)
y_test_logi = logi.predict(X_test)
print("model performance for test set: \n")
metric_score(y_test, y_test_logi)
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
val_scores = cross_val_score(logi, X_bal, y_bal, cv=5)  # 5-fold cross-validation

print("Logistic Regression Cross-Validation val_scores:", val_scores)
print("Mean Score:", val_scores.mean())
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100, max_depth=5)
rfc.fit(X_bal, y_bal)
y_pred_rfc=rfc.predict(X_bal)
print("model performance for training set: \n")
metric_score(y_bal, y_pred_rfc)
y_test_rfc = rfc.predict(X_test)
print("model performance for test set: \n")
metric_score(y_test, y_test_rfc)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_bal,y_bal)
y_pred_xgb=xgb.predict(X_bal)
print("model performance for training set: \n")
metric_score(y_bal, y_pred_xgb)
y_test_xgb = xgb.predict(X_test)
print("model performance for test set: \n")
metric_score(y_test, y_test_xgb)
```
